```python
import pandas as pd

df=pd.read_csv("../content/Superstore.csv",encoding="latin1")

df['Sales'].describe()

```

```python
df['Profit'].describe()
```

```python
top_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print(top_sales)

region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales.head)


df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
```

```python
top_sales = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print(top_sales.head)

region_sales = df.groupby('Region')['Profit'].sum()
print(region_sales.head)


df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)

```
